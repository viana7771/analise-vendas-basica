```python
import pandas as pd 
from matplotlib import pyplot as plt

# Carregar os dados
df = pd.read_csv(r'C:\Users\kaire\OneDrive\Área de Trabalho\Estudos de Dados\analise-vendas-ficticia\data\vendas.csv')
df.head()
```

```python
# Análise: total faturado
df['total'] = df['preco'] * df['quantidade']
faturamento_total = df['total'].sum()
print(f'Faturamento Total: R$ {faturamento_total:.2f}')
```

```python
# Análise: faturamento por produto
faturamento_por_produto = df.groupby('produto')['total'].sum()
faturamento_por_produto.plot(kind='bar', title='Faturamento por Produto')
plt.ylabel('Faturamento (R$)')
plt.show()
```

```python
# Insight: faturamento por produto

top_produto = df[df['produto'] == faturamento_por_produto.idxmax()]['quantidade'].sum()
print('Análise: Faturamento por produto:')
print(f'O produto que mais faturou foi o {faturamento_por_produto.idxmax()} \nCom um faturamento de R$ {faturamento_por_produto.max():.2f} \nCom o total de {top_produto} vendas.')
```

```python
# Análise: faturamento por categoria
faturamento_por_categoria = df.groupby('categoria')['total'].sum()
faturamento_por_categoria.plot(kind='bar', title='Faturamento por Categoria')
plt.ylabel('Faturamento (R$)')
plt.show()
```

```python
# Análise: em média, quanto cada venda gera de receita
media_faturamento = df['total'].mean()
print(f'Média de Faturamento por Venda: R$ {media_faturamento:.2f}')

# Análise: valor medio por transação por categoria
media_por_categoria = df.groupby('categoria')['total'].mean()
print(f'Média de Faturamento por transação por Categoria:')
print(media_por_categoria)
```
